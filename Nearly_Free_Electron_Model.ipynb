```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# PARAMETERS
N = 10 # Number of atoms in lattice

# PERIODIC POTENTIAL
# math.floor Round numbers down to the nearest integer
x = np.linspace(-0.1, 9.9, 1000)
y = np.array([1 if math.floor(2 * t) % 2 == 0 else 0 for t in x])

# Atoms in 1D lattice

x_atom = np.array([])
y_atom = np.array([])

for i in range(N):
   x_atom = np.append(x_atom,i + 0.25)
   y_atom = np.append(y_atom, 1.0)
```

```python
plt.figure(figsize=(20,10))
plt.plot(x,y,color = 'blue')
plt.plot(x_atom,y_atom,'o',color = 'red')
plt.show()
```

```python
# Wrap the lattice in a circle in order to avoid deals with the edge

L = 10                     # Lenght of the circumference
R = np.repeat(L/2*np.pi,N) # Radius 
theta_atoms = np.array([])
for n in range(N):
    theta_atoms = np.append(theta_atoms,2*np.pi*n/L)

# setting the axes projection as polar
plt.figure(figsize=(20,10))
plt.axes(projection = 'polar')
plt.polar(theta_atoms,R,'o',color = 'red')
plt.show()
```
